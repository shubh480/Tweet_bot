Tweepy makes it easier to use the twitter streaming api by handling authentication, connection, creating and destroying the session, reading incoming messages, and partially routing messages.

#### Import Tweepy and Time module

```python
import tweepy
import time
```

### Authentication

Tweepy tries to make OAuth 1a as painless as possible for you. To begin the process we need to register our client application with Twitter. Create a new application and once you are done you should have your consumer key and secret. Keep these two handy, you’ll need them.

Unlike basic authentication, we must do the OAuth before we can start using the API. We must complete the following steps:

1. Get a request token from twitter
2. Redirect user to twitter.com to authorize our application
3. If using a callback, twitter will redirect the user to us. Otherwise the user must manually supply us with the verifier code.
4. Exchange the authorized request token for an access token.

```python
#arguments are (API key, API scret key) in the consumer api keys
auth = tweepy.OAuthHandler("##########################", "############################################")
```

#### Token key

The final step is exchanging the request token for an access token. 

The access token is the “key” for opening the Twitter API treasure box. Twitter currently does not expire the tokens, so the only time it would ever go invalid is if the user revokes our application access. To store the access token depends on your application. Basically you need to store 2 string values: access token key and access token secret:

```python
#arguments are (access token, access token secret) from the access token and access token secret  keys
auth.set_access_token("######################################", "########################################")
```

##### So now that we have our OAuthHandler equipped with an access token, we are ready for business:

```python
#Access the API and set rate limits to True to avoid being banned
api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
```

```python
user = api.me()
```

The Twitter API platform offers three tiers of search APIs:
1. Standard
2. Premium
3. Enterprise

We have used the standard API which allows data of past 7 days only.

Documentation: "The standard search API searches against a sampling of recent Tweets published in the past 7 days. Part of the 'public' set of APIs."

### Enter Keywords to search related posts

Also enter the maximum number of posts the user wants to like. 

```python
#Keyword to use when searchinng for tweets/username to be liked
search = input("Enter a hastag or username: ")
numTweets = 10
```

### Output

With the above entered data, Tweepy will automatically print the number of liked posts.

```python
for tweet in tweepy.Cursor(api.search, search).items(numTweets):
    try:
        print('Tweet Liked')
        tweet.favorite()
        time.sleep(10)
    except tweepy.TweepError as e:
        print(e.reason)
    except StopIteration:
        break
```

Functions provided for Tweepy are as follows:- 

Parameters :	
1. auth_handler – authentication handler to be used
2. host – general API host
3. search_host – search API host
4. cache – cache backend to use
5. api_root – general API path root
6. search_root – search API path root
7. retry_count – default number of retries to attempt when error occurs
8. retry_delay – number of seconds to wait between retries
9. retry_errors – which HTTP status codes to retry
10. timeout – The maximum amount of time to wait for a response from Twitter
11. parser – The object to use for parsing the response from Twitter
12. compression – Whether or not to use GZIP compression for requests
13. wait_on_rate_limit – Whether or not to automatically wait for rate limits to replenish
14. wait_on_rate_limit_notify – Whether or not to print a notification when Tweepy is waiting for rate limits to replenish
15. proxy – The full url to an HTTPS proxy to use for connecting to Twitter.
